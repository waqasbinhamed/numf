```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

np.random.seed(42)
```

## Canada COVID - split timeframe

```python
# 2020-01-22 -- 2020-08-08

df = pd.read_csv(f'data/cases_s1.csv')
M = df['cases'].to_numpy().reshape(-1, 1)
m, n = M.shape

# without regularization
_load0 = np.load('saved_output/canada_sp1_it200.npz')
W0 = _load0['W']
H0 = _load0['H']

_load1 = np.load('saved_output/canada_reg_sp1_it200.npz')
W1 = _load1['W']
H1 = _load1['H']

fig, axs = plt.subplots(1, 3, figsize=(20, 6))
axs[0].plot(M)
axs[0].plot(W0 @ H0)
axs[0].plot(W1 @ H1)
axs[0].legend(['og', 'no reg', 'reg'])

axs[1].plot(W0)
axs[2].plot(W1)

print('no reg', np.linalg.norm(M - W0 @ H0, 'fro') / np.linalg.norm(M, 'fro'))
print('reg', np.linalg.norm(M - W1 @ H1, 'fro') / np.linalg.norm(M, 'fro'))
```
